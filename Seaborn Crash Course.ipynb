```python
df = pd.read_excel('house_price.xlsx')
df
```

# Corelation between different columns

```python
sns.pairplot(df)
```

# Scatter Plot

```python
sns.scatterplot(x='Area', y='Price', data = df)
```

# Join Plot - It helps to visualize the histogram and scatter plot together

```python
sns.jointplot(x='Area', y='Price', data=df)
```

# lmplot - It helps to visualize the scatter plot with a linear regfession already applied on it

```python
sns.lmplot(x='Area', y='Price', data=df)
```

# Countplot : Histogram/Bar chart is used for Numerical data where Countplot is used for categorical data

```python
df = pd.read_csv('http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv')
```

```python
df.head(5)
```

```python
ax = sns.countplot(x='Mfr Name', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```
